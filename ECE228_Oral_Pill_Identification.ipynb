## Run training and evaluation
Sample configuration options that we use and can be modified are following: 
* `--appearance_network`: base network layer. Can be chosen from resnet18, resnet50, restnet101, resnet152, densenet121, densenet161, and some more ResNet and DenseNet architecture.
* `--pooling`: pooling layer. Can be chosen from CBP, MPNCOV, BCNN, GAvP.
* `--max_epochs`: by default it's 300.
* `--data_root_dir`: should be `../../ePillID_data` if following the instruction.
* ` --ce_w`: the weight for Softmax. Default 1.0 for both plain classification and multihead-metric loss.
* ` --arcface_w`: the weight for cosine softmax. Default 0 for plain classification and 0.1 for multihead-metric loss.
* ` --contrastive_w`: the weight for contrastive loss. Default 0 for plain classification and 1.0 for multihead-metric loss.
* ` --triplet_w`: the weight for triplet loss. Default 0 for plain classification and 1.0 for multihead-metric loss.
* ` --load_mod`: load a half-trained model and continue training

Other configuration options can be checked out in `src/arguments.py`
